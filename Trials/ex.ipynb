```python
df_old = pd.read_excel("/Users/aswathshakthi/Downloads/master-entry-all-2026-02-19.xlsx")
df_old
```

```python
df_new =  pd.read_csv("/Users/aswathshakthi/Downloads/Database_Code_Aswath/master-entry-all-2026-02-19_validated.csv")
df_new
```

```python
import pandas as pd

def district_beneficiary_stats(df):
    """
    Filters Beneficiary Type == 'District'
    Groups by Beneficiary Name
    Returns:
        - unique article count (Requested Item)
        - total quantity sum
    """

    df = df.copy()
    df.columns = df.columns.str.strip()

    result = (
        df[df["Beneficiary Type"].astype(str).str.strip() == "District"]
        .assign(
            **{
                "Beneficiary Name": lambda x: x["Beneficiary Name"].astype(str).str.strip(),
                "Requested Item": lambda x: x["Requested Item"].astype(str).str.strip(),
                "Quantity": lambda x: pd.to_numeric(x["Quantity"], errors="coerce").fillna(0),
            }
        )
        .groupby("Beneficiary Name", dropna=False)
        .agg(
            unique_articles=("Requested Item", "nunique"),
            qty_sum=("Quantity", "sum"),
        )
        .reset_index()
    )

    return result

old_stats = district_beneficiary_stats(df_old)
new_stats = district_beneficiary_stats(df_new)

old_stats
```

```python
new_stats
```

```python
# Merge on Beneficiary Name
comparison = old_stats.merge(
    new_stats,
    on="Beneficiary Name",
    how="outer",
    suffixes=("_Old", "_New")
)

# Fill missing values (if a beneficiary exists only in one file)
comparison[["unique_articles_Old", "qty_sum_Old",
            "unique_articles_New", "qty_sum_New"]] = \
comparison[["unique_articles_Old", "qty_sum_Old",
            "unique_articles_New", "qty_sum_New"]].fillna(0)

# Optional: add difference columns
comparison["articles_diff"] = comparison["unique_articles_New"] - comparison["unique_articles_Old"]
comparison["qty_diff"] = comparison["qty_sum_New"] - comparison["qty_sum_Old"]

comparison
```

```python
output_file = "district_beneficiary_parallel.xlsx"
comparison.to_excel(output_file, index=False)

print(f"Saved as {output_file}")
```
